```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import numpy as np
from selenium.webdriver.common.keys import Keys
from urllib.parse import quote

news_number = []
news_title = []  # 빈 리스트 생성
news_write_date = []
news_content = []
news_writer = []
news_like = []
news_section = []
news_name=[]
# 1 = 동아일보, 2 = 조선일보, 3 = 중앙일보, 4 = 한겨례
pagelist = 1
newslist = 1
```

```python
search_query = "코로나 백신"
encoded_query = quote(search_query)

print(encoded_query)

webdriver_path = "C:/Users/이현주/Desktop/AA/chromedriver_win32/chromedriver"
driver = webdriver.Chrome(webdriver_path)

# 동아일보
donga_url = f'https://search.naver.com/search.naver?where=news&query={encoded_query}&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds=2020.03.01&de=2023.07.07&docid=&related=0&mynews=1&office_type=1&office_section_code=1&news_office_checked=1020&nso=so%3Ar%2Cp%3Afrom20200301to20230707&is_sug_officeid=0'

#한겨례
han_url = "https://search.naver.com/search.naver?where=news&query={encoded_query}&sm=tab_opt&sort=2&photo=3&field=0&pd=3&ds=2020.03.01&de=2023.07.07&docid=&related=0&mynews=1&office_type=2&office_section_code=14&news_office_checked=1028&nso=so%3Ar%2Cp%3Afrom20200301to20230707&is_sug_officeid=0"

driver.get(han_url)



for i in range(2280):
    # 블로그 클릭하기
    time.sleep(1)
    
    news = driver.find_element_by_xpath(f'//*[@id="sp_nws{newslist}"]/div/div/div[1]/div[2]/a[2]')
    news.click()
    newslist += 1
   
    # 현재 스크랩한 뉴스 순서대로
    news_number.append(i)

    # 뉴스 로딩 잠시 기다리기
    time.sleep(3)

    # 뉴스 타이틀 저장
    overlays = '//*[@id="title_area"]'                                

    # 요소가 로드될 때까지 대기
    tit = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, overlays)))
    tit = driver.find_element(By.XPATH, overlays)          # title
    title = tit.text
    print(title)
    news_title.append(title)

    # 뉴스 작성자 저장
    overlays = '//*[@id="contents"]/div[2]/p/span'                                
    nick = driver.find_element_by_xpath(overlays)         # nickname
    nickname = nick.text
    print(title)
    news_writer.append(nickname)

    # 뉴스 작성일 저장
    overlays = '//*[@id="ct"]/div[1]/div[3]/div[1]/div/span'                                
    date = driver.find_element_by_xpath(overlays)         # datetime
    datetime = date.text
    print(title)
    news_write_date.append(datetime)

    #뉴스 컨텐츠 저장
    overlays = '//*[@id="dic_area"]'                                
    contents = driver.find_element(By.XPATH ,overlays)    # contents
    cont = ""
    for content in contents:
        cont += content.text+"\n"
    news_content.append(cont)
    print(cont)

    # 뉴스 섹션 저장(사회면, 경제면 등등...)
    elements = driver.find_elements(By.XPATH, '//*[@id="contents"]/div[4]/a/em')
    print("요소 개수:", len(elements))
    
    for i in range(len(elements)):
        print(elements[i])
        if i != (len(elements)):
            news_section.append(elements[i]+"/")
            print(elements[i]+"/")
        else:
            news_section.append(elements[i])
            print(elements[i])
        
    # 크롤링이 성공하면 글 제목을 출력하게 되고,
    print(i, title)

    # 뉴스에 좋아요가 있다면 좋아요 수 저장, 없으면 0
    try:
        like = driver.find_element_by_css_selector('.u_likeit_text._count.num')
        print('like:', like.get_attribute("textContent"))
        like = int(like.get_attribute("textContent"))
        news_like.append(like)
    except:
        print('반응이 없습니다.')
        news_like.append('0')
    driver.close()

    # 원래 탭으로 다시 전환
    driver.switch_to.window(driver.window_handles[0])

    # 페이지 번호 누르기 1 2 3 4 5 ...
    if newslist == 10:
        pagelist += 1
        if pagelist % 10 != 0:
            page = driver.find_element_by_xpath(f'//*[@id="content"]/section/div[3]/span[{pagelist}]/a')
            page = driver.find_element_by_xpath(f'//*[@id="main_pack"]/div[2]/div/div/a[{pagelist}]')
            page.click()
        else:
            print("다음 페이지로 이동합니다.")
            page = driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/div/a[2]')
            page.click()
            pagelist = 1
           
        newslist = 1
        
```
